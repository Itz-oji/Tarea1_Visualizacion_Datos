```python
USER = "Itz-oji"
REPO = "Tarea1_Visualizacion_Datos"
BRANCH = "main"
BASE_URL = f"https://raw.githubusercontent.com/{USER}/{REPO}/{BRANCH}/Codigos/IA/"

print("Sincronizando datasets desde GitHub...")

# Load the datasets from GitHub, addressing previous FileNotFoundError
df_recaudaciones = pd.read_csv(BASE_URL + 'Recaudaciones.csv', encoding='latin1')
df_calificaciones = pd.read_excel(BASE_URL + 'Calificaciones.xlsx', engine='openpyxl')
df_tmdb = pd.read_csv(BASE_URL + 'tmdb_5000_movies.csv')
```

```python
print('--- Recaudaciones.csv ---')
display(df_recaudaciones.head())

print('\n--- Calificaciones(1).xlsx ---')
display(df_calificaciones.head())

print('\n--- tmdb_5000_movies.csv ---')
display(df_tmdb.head())
```

```python
# Clean 'RecaudaciÃ³n mundial' and 'Presupuesto' in df_recaudaciones
df_recaudaciones['RecaudaciÃ³n mundial'] = df_recaudaciones['RecaudaciÃ³n mundial'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).str.replace(' ', '', regex=False)
df_recaudaciones['RecaudaciÃ³n mundial'] = pd.to_numeric(df_recaudaciones['RecaudaciÃ³n mundial'], errors='coerce') / 1_000_000 # Convert to millions

df_recaudaciones['Presupuesto'] = df_recaudaciones['Presupuesto'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).str.replace(' ', '', regex=False)
df_recaudaciones['Presupuesto'] = pd.to_numeric(df_recaudaciones['Presupuesto'], errors='coerce') / 1_000_000 # Convert to millions

# Clean 'Nota CrÃ­tica (Normalizada)' in df_calificaciones
df_calificaciones['Nota CrÃ­tica'] = pd.to_numeric(df_calificaciones['Nota CrÃ­tica (Normalizada)'], errors='coerce')

# Extract primary genre from 'genres' in df_tmdb
import json
def get_primary_genre(genres_str):
    if isinstance(genres_str, str):
        try:
            genres_list = json.loads(genres_str)
            if genres_list:
                return genres_list[0]['name']
        except json.JSONDecodeError:
            pass
    return None

df_tmdb['Primary_Genre'] = df_tmdb['genres'].apply(get_primary_genre)

print('--- df_recaudaciones (cleaned) ---')
display(df_recaudaciones[['PelÃ­cula', 'RecaudaciÃ³n mundial', 'Presupuesto']].head())

print('\n--- df_calificaciones (cleaned) ---')
display(df_calificaciones[['PelÃ­cula', 'Nota CrÃ­tica']].head())

print('\n--- df_tmdb (cleaned) ---')
display(df_tmdb[['title', 'Primary_Genre']].head())
```

```python
# Fuzzy merge df_recaudaciones and df_calificaciones
def fuzzy_merge(df1, df2, key1, key2, threshold=85, limit=1):
    # Ensure the key columns are string type to prevent TypeErrors with floats/NaNs
    s = df2[key2].astype(str).tolist()
    m = df1[key1].astype(str).apply(lambda x: process.extract(x, s, limit=limit))
    df1['matches'] = m
    df1['match_score'] = df1['matches'].apply(lambda x: x[0][1] if x and x[0][1] >= threshold else None)
    df1['matched_title'] = df1['matches'].apply(lambda x: x[0][0] if x and x[0][1] >= threshold else None)
    return df1.drop(columns=['matches'])

df_merged_recaudaciones_calificaciones = fuzzy_merge(df_recaudaciones, df_calificaciones, 'PelÃ­cula', 'PelÃ­cula')

# Merge based on matched_title, bringing in 'Nota CrÃ­tica' from df_calificaciones
df_temp = df_merged_recaudaciones_calificaciones.merge(
    df_calificaciones[['PelÃ­cula', 'Nota CrÃ­tica']], # Select only 'PelÃ­cula' and 'Nota CrÃ­tica'
    left_on='matched_title',
    right_on='PelÃ­cula',
    how='left',
    suffixes=('_recaudaciones', '_calificaciones')
)

# Fuzzy merge with df_tmdb
df_temp = fuzzy_merge(df_temp, df_tmdb, 'PelÃ­cula_recaudaciones', 'title')

# Final merge with df_tmdb to get genre and budget
df_final = df_temp.merge(df_tmdb[['title', 'Primary_Genre', 'budget']], left_on='matched_title', right_on='title', how='left', suffixes=('_combined', '_tmdb'))

# Drop redundant columns and rename for clarity
df_final.drop(columns=['PelÃ­cula_calificaciones', 'title_tmdb'], inplace=True, errors='ignore')
df_final.rename(columns={'PelÃ­cula_recaudaciones': 'PelÃ­cula', 'matched_title': 'Matched_Movie_Title'}, inplace=True)

# Remove rows with NaN in critical columns after merge
df_final.dropna(subset=['RecaudaciÃ³n mundial', 'Nota CrÃ­tica', 'Presupuesto', 'Primary_Genre'], inplace=True)

print('--- Final Merged DataFrame (first 5 rows) ---')
display(df_final.head())
```

```python
# Calculate means for quadrant lines
mean_nota_critica = df_final['Nota CrÃ­tica'].mean()
mean_recaudacion = df_final['RecaudaciÃ³n mundial'].mean()

# Identify 5 most atypical films (outliers) based on distance from the mean center
# Calculate a 'distance' from the center of the quadrant for outlier detection
df_final['distance_from_center'] = np.sqrt(
    (df_final['Nota CrÃ­tica'] - mean_nota_critica)**2 +
    (df_final['RecaudaciÃ³n mundial'] - mean_recaudacion)**2
)

outlier_films = df_final.nlargest(5, 'distance_from_center')

# Create the quadrant chart using Plotly
fig = px.scatter(
    df_final,
    x='Nota CrÃ­tica',
    y='RecaudaciÃ³n mundial',
    color='Primary_Genre',
    size='Presupuesto',
    hover_name='PelÃ­cula',
    hover_data={'Nota CrÃ­tica': ':.1f', 'RecaudaciÃ³n mundial': ':.2f', 'Presupuesto': ':.2f', 'Primary_Genre': True},
    title='ClasificaciÃ³n de PelÃ­culas por Cuadrantes (Ãxito CrÃ­tico vs. Comercial)',
    labels={
        'Nota CrÃ­tica': 'Nota CrÃ­tica',
        'RecaudaciÃ³n mundial': 'RecaudaciÃ³n Mundial (millones USD)',
        'Primary_Genre': 'GÃ©nero Principal',
        'Presupuesto': 'Presupuesto (millones USD)'
    }
)

# Add quadrant lines
fig.add_shape(
    type='line',
    x0=mean_nota_critica, y0=df_final['RecaudaciÃ³n mundial'].min(),
    x1=mean_nota_critica, y1=df_final['RecaudaciÃ³n mundial'].max(),
    line=dict(color='Grey', width=1, dash='dash'),
    name=f'Mean Critical Score: {mean_nota_critica:.2f}'
)
fig.add_shape(
    type='line',
    x0=df_final['Nota CrÃ­tica'].min(), y0=mean_recaudacion,
    x1=df_final['Nota CrÃ­tica'].max(), y1=mean_recaudacion,
    line=dict(color='Grey', width=1, dash='dash'),
    name=f'Mean Worldwide Gross: {mean_recaudacion:.2f}'
)

# Add quadrant labels
# Define positions for annotations (adjust as needed)
x_range = df_final['Nota CrÃ­tica'].max() - df_final['Nota CrÃ­tica'].min()
y_range = df_final['RecaudaciÃ³n mundial'].max() - df_final['RecaudaciÃ³n mundial'].min()

fig.add_annotation(
    x=df_final['Nota CrÃ­tica'].max() - x_range * 0.1,
    y=df_final['RecaudaciÃ³n mundial'].max() - y_range * 0.1,
    text='Blockbuster CrÃ­tico',
    showarrow=False,
    font=dict(size=12, color='darkblue'),
    xanchor='right', yanchor='top'
)
fig.add_annotation(
    x=df_final['Nota CrÃ­tica'].min() + x_range * 0.1,
    y=df_final['RecaudaciÃ³n mundial'].max() - y_range * 0.1,
    text='Ãxito Comercial/Pop',
    showarrow=False,
    font=dict(size=12, color='darkgreen'),
    xanchor='left', yanchor='top'
)
fig.add_annotation(
    x=df_final['Nota CrÃ­tica'].max() - x_range * 0.1,
    y=df_final['RecaudaciÃ³n mundial'].min() + y_range * 0.1,
    text='Joyas de Culto',
    showarrow=False,
    font=dict(size=12, color='darkred'),
    xanchor='right', yanchor='bottom'
)
fig.add_annotation(
    x=df_final['Nota CrÃ­tica'].min() + x_range * 0.1,
    y=df_final['RecaudaciÃ³n mundial'].min() + y_range * 0.1,
    text='Bajo Impacto',
    showarrow=False,
    font=dict(size=12, color='darkorange'),
    xanchor='left', yanchor='bottom'
)

# Add labels for outlier films
for index, row in outlier_films.iterrows():
    fig.add_annotation(
        x=row['Nota CrÃ­tica'],
        y=row['RecaudaciÃ³n mundial'],
        text=row['PelÃ­cula'],
        showarrow=True,
        arrowhead=1,
        ax=0,
        ay=-20,
        font=dict(size=9, color='black')
    )

fig.update_layout(
    showlegend=True,
    xaxis=dict(range=[0, 10]),  # Set custom range for 'Nota CrÃ­tica' (x-axis)
    yaxis=dict(range=[0, 3000]) # Set custom range for 'RecaudaciÃ³n mundial' (y-axis)
)
fig.show()
```
